```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
from numpy import mat
```

```python
M=np.zeros((250,150))
M[:31,:]=1
M[:,60:91]=1
M[-31:,:60]=1
M[150:,:33]=1
plt.gray()
plt.imshow(M)
```

```python
M.shape
```

```python
#Adding noise to M plotting

M_noise=np.asmatrix(np.random.uniform(low=0,high=1,size=(250,150)))
plt.subplot(1,2,1)
plt.imshow(M_noise)
M_noise1=M+M_noise
plt.subplot(1,2,2)
plt.imshow(M_noise1)
```

```python
U,s,V=svd(M_noise1)
print(U.shape,s.shape,V.shape)

```

```python
for S in range(1,5):
    sigma=mat(np.eye(S)*s[:S])#eye for identitiy matrix
    print(sigma)
    print()
    u_reduced=U[:,:S]
    v_reduced=V[:S,:]
    M_reconstruct=u_reduced*sigma*v_reduced
    plt.subplot(1,5,S+1)
    plt.title('S=%d'%S)
    plt.imshow(M_reconstruct)
```
