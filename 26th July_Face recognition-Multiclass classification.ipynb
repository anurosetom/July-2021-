```python
from os import walk,path
#walk ,path helps to take directory name ,path and file name
import mahotas as mh
import numpy as np
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split,cross_val_score

from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
10304*400
```

```python
#Here our aim is to recognize the face of the person
#so here x is the 10 images in s1 folder, and y is s1 FOLDER 
```

```python
x=[]#image's path,ie image,which is x
y=[]
import warnings
warnings.filterwarnings('ignore')

for dir_path,dir_names,file_names in walk('/home/anu/Downloads/att_faces'):
    #printing file names,directory path
    #print(dir_path)
    for fn in file_names:
        if fn[-3:]=='pgm':#EXTRACTING ALL PGM EXTENSION FILES,-3 is for extract it using string format
            image_filename=path.join(dir_path,fn)
            print(image_filename)
            x.append(scale(mh.imread(image_filename,as_grey=True).reshape(10304).astype('float32')))
            #here this scale is for minmax scaling the image between 0 & 1, and read it as a grey scale image
            #and then convert 2d into 1d and convert into floating point
            y.append(dir_path)
    
    #apply this for entire 40 folders
```

```python
x=np.array(x)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=42)

pca=PCA(n_components=200)
xtrain_reduced=pca.fit_transform(xtrain)
xtest_reduced=pca.transform(xtest)

print('Original dimension of the training data',xtrain.shape)
print('reduced dimension of the training data',xtrain_reduced.shape)
```

```python
#using multi_class when we have multi class classification
model=LogisticRegression(multi_class='ovr',random_state=42)
model.fit(xtrain_reduced,ytrain)
```

```python
#checking the cross validation score

cross_valdn=cross_val_score(model,xtrain_reduced,ytrain)
print('Cross validation scores:',np.mean(cross_valdn),cross_valdn)
```

```python
predictions=model.predict(xtest_reduced)
print(classification_report(ytest,predictions))
```
