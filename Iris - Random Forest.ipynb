```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
iris
```

```python
df=pd.DataFrame(iris.data)
df
```

```python
df['target_names']=iris.target
```

```python
df[0:12]
```

```python
X=df.drop('target_names',axis='columns')
y=df.target_names
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=20)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(6,5))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree#to draw a classification
from sklearn import tree
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,random_state=42)
clf_dt=DecisionTreeClassifier(random_state=42)
clf_dt=clf_dt.fit(xtrain,ytrain)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(15,8),dpi=300)
a=tree.plot_tree(clf_dt,feature_names=X.columns,\
                 max_depth=None,filled=True)#also try max_depth=3 or None
fig.savefig('newtree.png') 
```
