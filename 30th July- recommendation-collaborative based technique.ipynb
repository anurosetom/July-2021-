```python
import pandas as pd
import sys
import numpy as np
```

```python
input_data=pd.read_csv('/home/anu/Downloads/movie.csv',index_col=0,na_values=['?'])
input_data
input_data.fillna(input_data.mean(),inplace=True)
input_data
```

```python
def movie_recom(user,movie,input_data,S=2):
    try:
        uindx=input_data.index.get_loc(user)
        print(uindx)
        bindx=input_data.columns.get_loc(movie)
        print(bindx)
    except:
        print('User Does not exist',sys.exc_info()[0])
        sys.exit(1)
        
    U,s,V=np.linalg.svd(input_data)
    print(U)
    print()
    print(s)
    print()
    print(V)
    print()
    
    reduced_sigma=np.mat(np.eye(S)*s[:S])
    U_reduced=U[:,:S]
    V_reduced=V[:S,:]
    
    print("reduced matrices")
    print(U_reduced)
    print()
    print(reduced_sigma)
    print()
    print(V_reduced)
    print()
    
    print('matrix w.r.t input index')
    print(U_reduced[uindx,:])
    print()
    print(reduced_sigma)
    print()
    print(np.mat(V_reduced[:,bindx]).T)
    
    recommended=U_reduced[uindx,:]*reduced_sigma*np.mat(V_reduced[:,bindx]).T
    
    return recommended.item(0)

bob_val=movie_recom('Bob','The Martian',input_data)
print()
print('Bob Score: %.2f'%bob_val)
print()
```
