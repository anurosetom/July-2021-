```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
from matplotlib.image import imread

plt.rcParams['figure.figsize']=[16,8]
```

```python
A=imread('/home/anu/Downloads/dog.jpg')
A
print(A.shape)
x=np.mean(A,-1)
print(x.shape)
img=plt.imshow(x)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U,S,VT=np.linalg.svd(x,full_matrices=False)
print(S)

#full_matrices for economic ____
```

```python
S=np.diag(S)

j=0
for r in (5,20,100):
    Xapprox=U[:,:r]@S[0:r,:r]@VT[:r,:]#@ for matrix multiplication
    plt.figure(j+1)
    j+=1
    img=plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.title('r='+str(r))
    plt.show()
    
    
    #check when r=5(first 5 points ,first 20 , first 100 points
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))#diagonal elements taken,ie taken here 1500
#semilogy for taking logs
plt.title('Singular values')
plt.show()
```

```python
plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))#diagonal elements/sum of diagonal elements, 
#here first 200 values shows variation in graph, remaining are constant
plt.title('Singular values: cumulative sum')
plt.show()
```
